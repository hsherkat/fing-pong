```python
from scipy.io.wavfile import read, write
import matplotlib.pyplot as plt
```

```python
input_data = read("click.wav")
rate, audio = input_data
```

```python
plt.plot(audio)
plt.ylabel("Amplitude")
plt.xlabel("Time")
plt.show()
```

```python
plt.plot(audio[7000:8500])
```

```python
plt.plot(audio[7750:8500])
```

```python
audio[:7750] *= 0
```

```python
write('click_new.wav', rate, audio)
```

```python
audio_trunc = audio[7000:]
write('click_new2.wav', rate, audio_trunc)
```
